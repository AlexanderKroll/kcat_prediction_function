```python
file = open(join("..", "..", "Tin_data", "tmp_list.pkl"),'rb')
df = pickle.load(file)
```

```python
columns = ["column_" +str(k) for k in range(len(df[0]))]

df_Tin = pd.DataFrame(columns=columns)
for ind in range(len(df)):
    d = {}
    for i in range(len(df[0])):
        d["column_" + str(i)] = df[ind][i]
    df_Tin = df_Tin.append(d, ignore_index=True)
df_Tin
```

```python
df = pd.DataFrame(columns = ["ID", "substrate list", "product list", "substrate string", "product string",
                            "sequence"])

for ind in df_Tin.index:
    sub, pro = df_Tin["column_5"][ind], df_Tin["column_6"][ind]
    sub_str, pro_str = ';'.join(sub), ';'.join(pro)
    enzymes = df_Tin["column_7"][ind]
    ID = df_Tin["column_0"][ind]

    for enz in list(set(enzymes)):
        df = df.append({"ID" : ID, "substrate list" : sub, "product list" : pro, "substrate string" : sub_str,
                       "product string" : pro_str, "sequence" : enz}, ignore_index = True)
        df = df.append({"ID" : ID + "_backward", "substrate list" : pro, "product list" : sub, "substrate string" : pro_str,
                       "product string" : sub_str, "sequence" : enz}, ignore_index = True)
```

```python
df_pred = kcat_predicton(substrates = list(df["substrate string"]),
                   products =  list(df["product string"]),
                   enzymes = list(df["sequence"])
                   )
```
